```python
# Data handling
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
train_df = pd.read_csv("../data/raw/train.csv")
test_df = pd.read_csv("../data/raw/test.csv")

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

train_df.head()
```

```python
# Count of those who survived as a percentage in train data
train_df['Survived'].value_counts(normalize=True)
```

```python
# Comparing sex and survived in train data
train_df.groupby('Sex')['Survived'].mean()
```

```python
# Comparing pclass and survived in train data
train_df.groupby('Pclass')['Survived'].mean()
```

```python
# Countplot of survived against sex
sns.countplot(data=train_df,x='Survived', hue='Sex')
plt.title('Survived against sex')
plt.show()
```

```python
# Countplot of survived against pclass
sns.countplot(data=train_df,x='Pclass',hue='Survived')
plt.title('Survived against Pclass ')
plt.show()
```

```python
# Histogram of survived against age
sns.histplot(data=train_df,x='Age',hue='Survived',multiple='stack',kde=True)
plt.title('Survived against Age')
plt.show()
```

```python
# Survival against sibsp
sns.countplot(data=train_df,x='SibSp',hue='Survived')
plt.title('Survived against SibSp')
plt.show()
```

```python
# Survival against parch
sns.countplot(data=train_df,x='Parch',hue='Survived')
plt.title('Survived against Parch')
plt.show()
```

```python
# Survival against embarked
sns.countplot(data=train_df,x='Embarked',hue='Survived')
plt.title('Survived against Embarked')
plt.show()
```

```python
# Histogram of survived against fare
sns.histplot(data=train_df,x='Fare',hue='Survived',multiple='stack',kde=True)
plt.title('Survived against Fare')
plt.show()
```

```python
# Correlaton matrix
corr=train_df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(10,7))
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation matrix')
plt.show
```
